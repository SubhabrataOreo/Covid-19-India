```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import datetime as dt
```

```python
data= pd.read_csv('Documents/Covid 19 states India.csv', parse_dates=['Date'], dayfirst=True)
```

```python
data.head(10)
```

```python
data=data[['Date','State/UnionTerritory','Cured','Deaths','Confirmed']]
```

```python
data.head(10)
```

```python
data.columns=['date','state','cured','deaths','confirmed']
```

```python
data.head()
```

```python
data.tail()
```

```python
today=data[data.date=='2020-07-18']
```

```python
today
```

```python
today_confirmed_asc=today.sort_values("confirmed", ascending=False)
```

```python
today_confirmed_asc
```

```python
today_confirmed_max= today_confirmed_asc[0:5]
```

```python
today_confirmed_max
```

```python
sns.set(rc={'figure.figsize':(15,10)})
sns.barplot(x="state",y="confirmed",data=today_confirmed_max, hue="state")
plt.title('Maximum confirmed today')
plt.show()
```

```python
today_death_asc=today.sort_values("deaths", ascending=False)
```

```python
today_death_asc
```

```python
today_death_max=today_death_asc[0:5]
sns.set(rc={"figure.figsize":(15,10)})
sns.barplot(x="state",y="deaths",data=today_death_max, hue="state")
plt.title("Today Maximum Deaths")
plt.show()
```

```python
today_cured_asc=today.sort_values("cured", ascending=False)
today_cured_asc
```

```python
today_cured_max=today_cured_asc[0:5]
sns.set(rc={"figure.figsize":(15,10)})
sns.barplot(x="state",y="cured",data=today_cured_max, hue="state")
plt.show()
```

```python
WB= data[data.state=='West Bengal']
```

```python
WB
```

```python
sns.set(rc={'figure.figsize':(15,10)})
sns.lineplot(x="date", y="confirmed",data=WB,color="b")
plt.title("WB_Confirmed")
plt.show()
```

```python
sns.set(rc={'figure.figsize':(15,10)})
sns.lineplot(x="date", y="deaths",data=WB,color="r")
plt.title("WB_Deaths")
plt.show()
```

```python
WB_specific=WB[WB['date'].between('2020-05-01','2020-05-10',inclusive=True)]
```

```python
sns.set(rc={'figure.figsize':(15,10)})
sns.lineplot(x="date", y="deaths",data=WB_specific,color="g")
plt.title("WB_Deaths_specific")
plt.show()
```
